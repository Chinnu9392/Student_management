```python
class Student:

    def __init__(self, id, name):
        self.id = id
        self.name = name
        self.courses = []
        self.grades = {}

    def enroll(self, course):
        if course not in self.courses:
            self.courses.append(course.name)
            course.add_student(self)
            
    def get_grade(self, course):
        return self.grades.get(course)

    def get_gpa(self):
        total_points = 0
        credits = 0
        for course, grade in self.grades.items():  
            #declaring the student As failed if he get 0 in anyone of the courses
            if grade.value==0:
                return "Fail"
                break
            if grade.value >4:
                points = grade.value * course.credits
                total_points += points
                credits += course.credits
            
        if credits > 0:
            return total_points / credits
        else:
            return None
        
    def get_enrolled_courses(self):
        return self.courses

class Course:

    def __init__(self, id, name, credits):
        self.id = id
        self.name = name
        self.credits = credits
        self.students = []

    def add_student(self, student):
        self.students.append(student.name)

    def get_enrolled_students(self):
        return self.students

class Grade:

    def __init__(self, marks):
        self.marks=marks
        self.value = self.get_numeric_grade()
        
    def get_numeric_grade(self):
        if self.marks>=90:
            return 10
        elif self.marks>=80:
            return 9
        elif self.marks>=70:
            return 8
        elif self.marks >=60:
            return 7
        elif self.marks>=50:
            return 6
        elif self.marks>=40:
            return 5
        elif self.marks<40:
            return 0

# Example usage
student1 = Student(1, "Sai")
student2 = Student(2, "Ram")
course1 = Course(101, "Introduction to Programming", 3)
course2 = Course(102, "Introduction to Python", 3)
course3 = Course(103, "JAVA", 3)


student1.enroll(course1)
student1.enroll(course2)
student2.enroll(course1)
student2.enroll(course3)

grade1 = Grade(84)
student1.grades[course1] = grade1
grade2 = Grade(95)
student1.grades[course2] = grade2

print(student1.get_grade(course1).value)
print((course1.get_enrolled_students()))

print((student1.get_enrolled_courses()))
print((student2.get_enrolled_courses()))
print(f"Student {student1.name}'s GPA: {student1.get_gpa()}")
```
